```python
import os
import numpy as np
import json
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import clip
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1. Preprocess Data (Normalization/Standardization)
def normalize_images(images):
    return images / 255.0

def standardize_fmri(fmri_data):
    mean = np.mean(fmri_data, axis=0)
    std = np.std(fmri_data, axis=0)
    return (fmri_data - mean) / std

def preprocess_and_save(subject_num, raw_data_dir, preprocessed_data_dir):
    image_data = np.load(os.path.join(raw_data_dir, 'images_256.npz'))
    train_images = normalize_images(image_data['train_images'])
    test_images = normalize_images(image_data['test_images'])

    subject_data = np.load(os.path.join(raw_data_dir, f'sbj_{subject_num}.npz'))
    train_fmri = standardize_fmri(subject_data['arr_0'])
    test_fmri = standardize_fmri(subject_data['arr_2'])

    np.savez(os.path.join(preprocessed_data_dir, f'subject_{subject_num}_train.npz'), 
             images=train_images, fmri=train_fmri)
    np.savez(os.path.join(preprocessed_data_dir, f'subject_{subject_num}_test.npz'), 
             images=test_images, fmri=test_fmri)

# Paths
raw_data_dir = '/kaggle/working/dataset/data/Kamitani/npz'
preprocessed_data_dir = '/kaggle/working/preprocessed'
os.makedirs(preprocessed_data_dir, exist_ok=True)
for subject_num in range(1, 6):
    preprocess_and_save(subject_num, raw_data_dir, preprocessed_data_dir)

# 2. Dataset Class
class FMRITextDataset(Dataset):
    def __init__(self, fmri_file, text_embeddings):
        data = np.load(fmri_file)
        self.fmri = data['fmri']
        self.text_embeddings = text_embeddings

    def __len__(self):
        return len(self.fmri)

    def __getitem__(self, idx):
        fmri = torch.tensor(self.fmri[idx], dtype=torch.float32)
        text_embedding = torch.tensor(self.text_embeddings[idx], dtype=torch.float32)
        return fmri, text_embedding

# 3. Model Definition
class FMRIToTextModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(FMRIToTextModel, self).__init__()
        self.mlp = nn.Sequential(
            nn.Linear(input_dim, 1024), nn.BatchNorm1d(1024), nn.ReLU(),
            nn.Linear(1024, 768), nn.BatchNorm1d(768), nn.ReLU(),
            nn.Linear(768, output_dim)
        )

    def forward(self, x):
        return self.mlp(x)

# 4. Cosine Similarity Loss + MSE
class CosineMSELoss(nn.Module):
    def __init__(self, lambda_cosine=0.5):
        super().__init__()
        self.lambda_cosine = lambda_cosine
        self.mse = nn.MSELoss()

    def forward(self, predicted, target):
        mse_loss = self.mse(predicted, target)
        cosine_loss = 1 - nn.functional.cosine_similarity(predicted, target).mean()
        return mse_loss + self.lambda_cosine * cosine_loss

# 5. Load CLIP-ViT-Large-14 Model
clip_model, preprocess = clip.load("ViT-L/14", device)

def extract_text_embeddings(text_descriptions):
    tokens = clip.tokenize(text_descriptions).to(device)
    with torch.no_grad():
        return clip_model.encode_text(tokens).cpu().numpy()

# 6. Train Model
def train_model(train_loader, model, criterion, optimizer, num_epochs=500):
    model.train()
    for epoch in range(num_epochs):
        total_loss = 0
        for fmri, text_embedding in train_loader:
            fmri, text_embedding = fmri.to(device), text_embedding.to(device)
            optimizer.zero_grad()
            predicted = model(fmri)
            loss = criterion(predicted, text_embedding)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}")

# 7. Main Pipeline
train_file = f'{preprocessed_data_dir}/subject_1_train.npz'
test_file = f'{preprocessed_data_dir}/subject_1_test.npz'

# Load captions and extract embeddings
with open('/kaggle/input/god-caps/GOD_train_caps_hm_and_wdnet.json', 'r') as f:
    captions_data = json.load(f)

keys = list(captions_data.keys())
text_descriptions = [captions_data[key][1] for key in keys[:1200]]
text_embeddings = extract_text_embeddings(text_descriptions)

# Prepare Dataset
train_dataset = FMRITextDataset(train_file, text_embeddings)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Train FMRI -> Text Embeddings Model
model = FMRIToTextModel(input_dim=4466, output_dim=768).to(device)
criterion = CosineMSELoss(lambda_cosine=0.5)
optimizer = optim.Adam(model.parameters(), lr=0.0005)

train_model(train_loader, model, criterion, optimizer)
torch.save(model.state_dict(), '/kaggle/working/fmri_to_text_model.pth')

# 8. Generate Images with Stable Diffusion
model.eval()
test_data = np.load(test_file)
test_fmri = torch.tensor(test_data['fmri'], dtype=torch.float32).to(device)

# Predict Text Embeddings
with torch.no_grad():
    predicted_text_embeddings = model(test_fmri)

# Load Stable Diffusion Pipeline
pipeline = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to(device)

# Generate Images
def generate_images(predicted_embeddings, images, num_samples=5):
    for i in range(num_samples):
        embedding = predicted_embeddings[i].unsqueeze(0)
        generated_image = pipeline(prompt_embeds=embedding).images[0]

        # Display images
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))
        axes[0].imshow((images[i] * 255).astype(np.uint8))
        axes[0].set_title("Original Image")
        axes[1].imshow(generated_image)
        axes[1].set_title("Generated Image")
        plt.show()

generate_images(predicted_text_embeddings, test_data['images'])
```

```python
def generate_images(predicted_embeddings, images, num_samples=5):
    for i in range(num_samples):
        # Reshape embeddings
        embedding = predicted_embeddings[i].unsqueeze(0).unsqueeze(1).to(device)
        negative_embedding = torch.zeros_like(embedding).to(device)  # Dummy negative prompt

        # Generate image with Stable Diffusion
        generated_image = pipeline(
            prompt_embeds=embedding, 
            negative_prompt_embeds=negative_embedding, 
            guidance_scale=7.5  # Adjust CFG scale (standard is ~7.5)
        ).images[0]

        # Display images
        fig, axes = plt.subplots(1, 2, figsize=(10, 5))
        axes[0].imshow((images[i] * 255).astype(np.uint8))
        axes[0].set_title("Original Image")
        axes[1].imshow(generated_image)
        axes[1].set_title("Generated Image")
        plt.show()
```
