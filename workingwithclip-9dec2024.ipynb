```python
import numpy as np
import os

# Define a function to normalize image data
def normalize_images(images):
    return images / 255.0  # Normalize RGB values to [0, 1]

# Define a function to standardize fMRI data
def standardize_fmri(fmri_data):
    mean = np.mean(fmri_data, axis=0)
    std = np.std(fmri_data, axis=0)
    return (fmri_data - mean) / std  # Standardize to mean 0 and std 1

# Main function to preprocess data
def preprocess_and_save(subject_num, raw_data_dir, preprocessed_data_dir):
    # Load image data
    image_data = np.load(os.path.join(raw_data_dir, 'images_256.npz'))
    train_images = image_data['train_images']
    test_images = image_data['test_images']

    # Normalize images
    train_images_normalized = normalize_images(train_images)
    test_images_normalized = normalize_images(test_images)

    # Load fMRI data for the subject
    subject_data = np.load(os.path.join(raw_data_dir, f'sbj_{subject_num}.npz'))
    train_fmri = subject_data['arr_0']
    test_fmri = subject_data['arr_2']

    # Standardize fMRI data
    train_fmri_standardized = standardize_fmri(train_fmri)
    test_fmri_standardized = standardize_fmri(test_fmri)

    # Save the preprocessed data as .npz files in the preprocessed data directory
    np.savez(os.path.join(preprocessed_data_dir, f'preprocessed_data_subject_{subject_num}_train.npz'),
             images=train_images_normalized, fmri=train_fmri_standardized)
    np.savez(os.path.join(preprocessed_data_dir, f'preprocessed_data_subject_{subject_num}_test.npz'),
             images=test_images_normalized, fmri=test_fmri_standardized)

    print(f"Preprocessed data for Subject {subject_num} saved as .npz files.")

# Set paths for Colab
raw_data_dir = '/kaggle/working/dataset/data/Kamitani/npz'
preprocessed_data_dir = '/kaggle/working/dataset/preprocessed'

# Ensure the output directory exists
os.makedirs(preprocessed_data_dir, exist_ok=True)
# Preprocess and save data for each subject
for subject_num in range(1, 6):
    preprocess_and_save(subject_num, raw_data_dir, preprocessed_data_dir)
```

```python
import torch
import clip
from torchvision.transforms import Compose, Resize, CenterCrop, ToTensor, Normalize
import numpy as np
import os
from PIL import Image

# Step 1: Load the CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Step 2: Define a function to preprocess images
def preprocess_images(images):
    """Preprocess images for CLIP."""
    images = [Image.fromarray((img * 255).astype(np.uint8)) for img in images]  # Convert normalized images back to [0, 255]
    return torch.stack([preprocess(img).to(device) for img in images])

# Step 3: Extract image embeddings
def extract_image_embeddings(preprocessed_file, model):
    """Extract CLIP image embeddings from preprocessed files."""
    data = np.load(preprocessed_file)
    images = data["images"]  # Load normalized images
    preprocessed_images = preprocess_images(images)
    with torch.no_grad():
        image_features = model.encode_image(preprocessed_images)
    return image_features.cpu().numpy()

# Step 4: Save embeddings
def save_embeddings(embeddings, file_name):
    """Save embeddings as a NumPy file."""
    np.save(file_name, embeddings)
    print(f"Embeddings saved to {file_name}")

# Directory paths
preprocessed_data_dir = '/kaggle/working/dataset/preprocessed'
output_embeddings_dir = '/kaggle/working/dataset/embeddings'

# Ensure output directory exists
os.makedirs(output_embeddings_dir, exist_ok=True)

# Step 5: Extract and save image embeddings for each subject
for subject_num in range(1, 6):
    preprocessed_file = os.path.join(preprocessed_data_dir, f'preprocessed_data_subject_{subject_num}_train.npz')
    print(f"Processing Subject {subject_num}...")
    image_embeddings = extract_image_embeddings(preprocessed_file, model)
    save_embeddings(image_embeddings, os.path.join(output_embeddings_dir, f'subject_{subject_num}_image_embeddings.npy'))
```

```python
import json

# Step 1: Load captions
def load_captions(json_file):
    """Load captions from JSON file."""
    with open(json_file, "r") as f:
        data = json.load(f)
    captions = []
    for key, caption_list in data.items():
        captions.append(caption_list[0])  # Use the first caption per image
    return captions

# Step 2: Extract text embeddings
def extract_text_embeddings(captions, model):
    """Extract CLIP text embeddings."""
    with torch.no_grad():
        text_tokens = clip.tokenize(captions).to(device)
        text_features = model.encode_text(text_tokens)
    return text_features.cpu().numpy()

# Step 3: Save text embeddings
def save_embeddings(embeddings, file_name):
    """Save embeddings as a NumPy file."""
    np.save(file_name, embeddings)
    print(f"Embeddings saved to {file_name}")

# Path to captions JSON file
caption_json_path = '/kaggle/input/god-caps/GOD_train_caps_hm_and_wdnet.json'

# Load captions
captions = load_captions(caption_json_path)

# Extract text embeddings
print("Extracting text embeddings...")
text_embeddings = extract_text_embeddings(captions, model)

# Save text embeddings
output_file = '/kaggle/working/dataset/embeddings/text_embeddings.npy'
save_embeddings(text_embeddings, output_file)
```

```python
# Extract image embeddings for test data
def extract_test_image_embeddings(preprocessed_file, model):
    """Extract CLIP image embeddings for the test set."""
    data = np.load(preprocessed_file)
    test_images = data["images"]  # Load test images
    preprocessed_test_images = preprocess_images(test_images)
    with torch.no_grad():
        test_image_features = model.encode_image(preprocessed_test_images)
    return test_image_features.cpu().numpy()

# Path to preprocessed test data
test_data_file = '/kaggle/working/dataset/preprocessed/preprocessed_data_subject_1_test.npz'

# Extract and save test image embeddings
test_image_embeddings = extract_test_image_embeddings(test_data_file, model)
test_image_emb_path = '/kaggle/working/dataset/embeddings/subject_1_test_image_embeddings.npy'
save_embeddings(test_image_embeddings, test_image_emb_path)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import numpy as np
import os

# Define the dataset
class FMRIEmbeddingDataset(Dataset):
    def __init__(self, fmri_file, image_emb_file):
        # Load fMRI and image embeddings
        fmri_data = np.load(fmri_file)['fmri']
        image_embeddings = np.load(image_emb_file)
        
        # Ensure dimensions match
        assert fmri_data.shape[0] == image_embeddings.shape[0], "Mismatch in dataset sizes."
        
        self.fmri = torch.tensor(fmri_data, dtype=torch.float32)
        self.image_emb = torch.tensor(image_embeddings, dtype=torch.float32)

    def __len__(self):
        return len(self.fmri)

    def __getitem__(self, idx):
        return self.fmri[idx], self.image_emb[idx]

# Define the fMRI Encoder Model
class FMRIEncoder(nn.Module):
    def __init__(self, input_size=4466, output_size=512):
        super(FMRIEncoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_size, 1024),
            nn.ReLU(),
            nn.Linear(1024, 512)
        )

    def forward(self, x):
        return self.encoder(x)

# Load dataset paths
fmri_file = '/kaggle/working/dataset/preprocessed/preprocessed_data_subject_1_train.npz'
image_emb_file = '/kaggle/working/dataset/embeddings/subject_1_image_embeddings.npy'

# Load dataset and create DataLoader
dataset = FMRIEmbeddingDataset(fmri_file, image_emb_file)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Initialize model, loss, and optimizer
model = FMRIEncoder().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training Loop
num_epochs = 50
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for fmri, image_emb in train_loader:
        fmri = fmri.to(device)
        image_emb = image_emb.to(device)

        # Forward pass
        output = model(fmri)
        loss = criterion(output, image_emb)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(train_loader):.4f}")

# Save trained model
model_save_path = '/kaggle/working/dataset/embeddings/fmri_encoder.pth'
torch.save(model.state_dict(), model_save_path)
print(f"Model saved to {model_save_path}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Load the trained model
model.load_state_dict(torch.load(model_save_path))
model.eval()

# Load test data
test_data_file = '/kaggle/working/dataset/preprocessed/preprocessed_data_subject_1_test.npz'
test_data = np.load(test_data_file)
test_fmri = torch.tensor(test_data['fmri'], dtype=torch.float32).to(device)
original_image_emb = np.load('/kaggle/working/dataset/embeddings/subject_1_image_embeddings.npy')

# Generate embeddings using the trained fMRI encoder
with torch.no_grad():
    generated_emb = model(test_fmri).cpu().numpy()

# Calculate cosine similarity
similarities = cosine_similarity(generated_emb, original_image_emb)

# Find and display the best matches
def display_best_matches(similarities, original_images, num_matches=5):
    best_matches = np.argmax(similarities, axis=1)
    fig, axes = plt.subplots(num_matches, 2, figsize=(10, 20))

    for i in range(num_matches):
        original_img_idx = best_matches[i]

        # Load and display the images
        original_img = Image.fromarray((original_images[original_img_idx] * 255).astype(np.uint8))
        reconstructed_img = Image.fromarray((original_images[i] * 255).astype(np.uint8))

        axes[i, 0].imshow(reconstructed_img)
        axes[i, 0].set_title("Reconstructed from fMRI")
        axes[i, 0].axis("off")

        axes[i, 1].imshow(original_img)
        axes[i, 1].set_title("Closest CLIP Image")
        axes[i, 1].axis("off")

    plt.show()

# Load the original test images
test_images = test_data['images']

# Display top-5 matches
display_best_matches(similarities, test_images)
```
