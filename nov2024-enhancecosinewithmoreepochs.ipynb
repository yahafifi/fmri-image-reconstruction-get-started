```python
import numpy as np
import os
from sklearn.decomposition import PCA

# Define the directory paths
raw_data_dir = '/kaggle/working/dataset/data/Kamitani/npz'
preprocessed_data_dir = '/kaggle/working/dataset/preprocessed/fmri'
os.makedirs(preprocessed_data_dir, exist_ok=True)

# Function to standardize fMRI data
def standardize_fmri(fmri_data):
    mean = np.mean(fmri_data, axis=0)
    std = np.std(fmri_data, axis=0)
    return (fmri_data - mean) / std  # Standardize to mean 0 and std 1

# Function to apply PCA for dimensionality reduction
def reduce_dimensionality(fmri_data, variance_ratio=0.95):
    pca = PCA(n_components=variance_ratio)  # Retain 95% variance
    reduced_data = pca.fit_transform(fmri_data)
    print(f"PCA reduced dimensions to {reduced_data.shape[1]} components")
    return reduced_data, pca

# Function to preprocess fMRI data for one subject
def preprocess_fmri_subject(subject_num, raw_data_dir, preprocessed_data_dir, apply_pca=True):
    # Load fMRI data for the subject
    subject_file = os.path.join(raw_data_dir, f'sbj_{subject_num}.npz')
    fmri_data = np.load(subject_file)
    train_fmri = fmri_data['arr_0']  # Training fMRI data
    test_fmri = fmri_data['arr_2']   # Testing fMRI data

    # Standardize fMRI data
    train_fmri_standardized = standardize_fmri(train_fmri)
    test_fmri_standardized = standardize_fmri(test_fmri)

    # Optionally apply PCA for dimensionality reduction
    if apply_pca:
        train_fmri_reduced, pca = reduce_dimensionality(train_fmri_standardized, variance_ratio=0.95)
        test_fmri_reduced = pca.transform(test_fmri_standardized)  # Use the same PCA model for test data
    else:
        train_fmri_reduced = train_fmri_standardized
        test_fmri_reduced = test_fmri_standardized

    # Save preprocessed data
    np.savez(os.path.join(preprocessed_data_dir, f'preprocessed_fmri_subject_{subject_num}.npz'),
             train_fmri=train_fmri_reduced, test_fmri=test_fmri_reduced)
    print(f"Preprocessed fMRI data saved for Subject {subject_num}")

# Preprocess all subjects
for subject_num in range(1, 6):  # Assuming 5 subjects
    preprocess_fmri_subject(subject_num, raw_data_dir, preprocessed_data_dir)
```

```python
!pip install git+https://github.com/openai/CLIP.git

import numpy as np
import torch
from PIL import Image
from torchvision import transforms
import os
import clip

# Load the CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-L/14", device=device)

# Define the paths
raw_data_dir = '/kaggle/working/dataset/data/Kamitani/npz'
preprocessed_image_dir = '/kaggle/working/dataset/preprocessed/images'
os.makedirs(preprocessed_image_dir, exist_ok=True)

# Function to extract image embeddings using CLIP
def extract_image_embeddings(image_data, model, preprocess, device):
    embeddings = []
    for img_array in image_data:
        # Convert to PIL Image and preprocess for CLIP
        img = Image.fromarray((img_array * 255).astype('uint8'))  # Convert to 8-bit format
        img_preprocessed = preprocess(img).unsqueeze(0).to(device)  # Add batch dimension and move to device
        with torch.no_grad():
            img_embedding = model.encode_image(img_preprocessed)
            embeddings.append(img_embedding.cpu().numpy())
    return np.vstack(embeddings)  # Stack all embeddings

# Main function for preprocessing images
def preprocess_images(raw_data_dir, preprocessed_image_dir, model, preprocess, device):
    # Load the images from .npz
    image_data = np.load(os.path.join(raw_data_dir, 'images_256.npz'))
    train_images = image_data['train_images']
    test_images = image_data['test_images']

    # Extract embeddings
    print("Extracting embeddings for training images...")
    train_embeddings = extract_image_embeddings(train_images, model, preprocess, device)
    print("Extracting embeddings for testing images...")
    test_embeddings = extract_image_embeddings(test_images, model, preprocess, device)

    # Save preprocessed image embeddings
    np.savez(os.path.join(preprocessed_image_dir, 'preprocessed_image_embeddings.npz'),
             train_image_embeddings=train_embeddings, test_image_embeddings=test_embeddings)
    print(f"Image embeddings saved to {preprocessed_image_dir}")

# Run the preprocessing
preprocess_images(raw_data_dir, preprocessed_image_dir, model, preprocess, device)
```

```python
import json
import torch
import os
import numpy as np

# Load the CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-L/14", device=device)

# Define paths
raw_caption_file = '/kaggle/input/god-caps/GOD_train_caps_hm_and_wdnet.json'
preprocessed_caption_dir = '/kaggle/working/dataset/preprocessed/captions'
os.makedirs(preprocessed_caption_dir, exist_ok=True)

# Load captions from the JSON file
def load_captions(caption_file):
    with open(caption_file, 'r') as file:
        captions = json.load(file)
    return captions

# Extract text embeddings using CLIP
def extract_text_embeddings(captions, model, device):
    embeddings = []
    ids = []
    for image_id, caption in captions.items():
        text_input = clip.tokenize(caption).to(device)  # Tokenize text for CLIP
        with torch.no_grad():
            text_embedding = model.encode_text(text_input).cpu().numpy()  # Extract text embedding
        embeddings.append(text_embedding)
        ids.append(image_id)
    return np.vstack(embeddings), ids

# Main function to preprocess captions
def preprocess_captions(raw_caption_file, preprocessed_caption_dir, model, device):
    # Load the captions
    captions = load_captions(raw_caption_file)
    print(f"Loaded {len(captions)} captions")

    # Extract text embeddings
    print("Extracting embeddings for captions...")
    text_embeddings, image_ids = extract_text_embeddings(captions, model, device)

    # Save the embeddings and IDs
    np.savez(os.path.join(preprocessed_caption_dir, 'preprocessed_caption_embeddings.npz'),
             text_embeddings=text_embeddings, image_ids=image_ids)
    print(f"Caption embeddings saved to {preprocessed_caption_dir}")

# Run the preprocessing
preprocess_captions(raw_caption_file, preprocessed_caption_dir, model, device)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from transformers import BertTokenizer, BertModel
```

```python
class ImageEncoder(nn.Module):
    def __init__(self, embed_dim):
        super(ImageEncoder, self).__init__()
        # Use a pretrained ResNet as a feature extractor
        self.cnn = models.resnet50(pretrained=True)
        self.cnn.fc = nn.Identity()  # Remove the final classification layer
        self.projection = nn.Linear(2048, embed_dim)  # Map to shared latent space

    def forward(self, images):
        features = self.cnn(images)  # Extract features
        return self.projection(features)  # Project to shared latent space
```

```python
class TextEncoder(nn.Module):
    def __init__(self, embed_dim):
        super(TextEncoder, self).__init__()
        # Use a pretrained BERT model
        self.bert = BertModel.from_pretrained('bert-base-uncased')
        self.projection = nn.Linear(768, embed_dim)  # Map to shared latent space

    def forward(self, input_ids, attention_mask):
        # Extract features from the BERT model
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        features = outputs.last_hidden_state[:, 0, :]  # Use the CLS token embedding
        return self.projection(features)  # Project to shared latent space
```

```python
class CustomDataset(Dataset):
    def __init__(self, image_paths, captions, tokenizer, transform):
        self.image_paths = image_paths
        self.captions = captions
        self.tokenizer = tokenizer
        self.transform = transform

    def __len__(self):
        return len(self.captions)

    def __getitem__(self, idx):
        # Load image
        image = Image.open(self.image_paths[idx]).convert("RGB")
        image = self.transform(image)

        # Tokenize caption
        caption = self.captions[idx]
        encoded = self.tokenizer(caption, padding="max_length", truncation=True, max_length=30, return_tensors="pt")
        return image, encoded['input_ids'].squeeze(0), encoded['attention_mask'].squeeze(0)
```

```python
class CLIPLoss(nn.Module):
    def __init__(self, temperature=0.1):
        super(CLIPLoss, self).__init__()
        self.temperature = temperature

    def forward(self, image_embeddings, text_embeddings):
        # Normalize embeddings
        image_embeddings = nn.functional.normalize(image_embeddings, dim=1)
        text_embeddings = nn.functional.normalize(text_embeddings, dim=1)

        # Compute similarity
        logits = torch.matmul(image_embeddings, text_embeddings.T) / self.temperature

        # Labels for positive pairs
        labels = torch.arange(len(image_embeddings)).to(image_embeddings.device)

        # Cross-entropy loss
        loss = nn.CrossEntropyLoss()
        return loss(logits, labels) + loss(logits.T, labels)  # Symmetric loss
```

```python
# Hyperparameters
embed_dim = 256
batch_size = 32
num_epochs = 10
learning_rate = 1e-4

# Prepare the dataset and dataloader
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

dataset = CustomDataset(image_paths, captions, tokenizer, transform)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Initialize models, optimizer, and loss
image_encoder = ImageEncoder(embed_dim).to('cuda')
text_encoder = TextEncoder(embed_dim).to('cuda')
criterion = CLIPLoss().to('cuda')
optimizer = optim.Adam(list(image_encoder.parameters()) + list(text_encoder.parameters()), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    image_encoder.train()
    text_encoder.train()

    total_loss = 0
    for images, input_ids, attention_mask in dataloader:
        images = images.to('cuda')
        input_ids = input_ids.to('cuda')
        attention_mask = attention_mask.to('cuda')

        # Forward pass
        image_embeddings = image_encoder(images)
        text_embeddings = text_encoder(input_ids, attention_mask)
        loss = criterion(image_embeddings, text_embeddings)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(dataloader):.4f}")
```

```python
import os
import glob
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
```

```python
class ImageEncoder(nn.Module):
    def __init__(self, embed_dim):
        super(ImageEncoder, self).__init__()
        self.cnn = models.resnet50(pretrained=True)
        self.cnn.fc = nn.Identity()  # Remove the classification layer
        self.projection = nn.Linear(2048, embed_dim)  # Project to shared latent space

    def forward(self, images):
        features = self.cnn(images)
        return self.projection(features)
```

```python
class TextEncoder(nn.Module):
    def __init__(self, embed_dim):
        super(TextEncoder, self).__init__()
        self.bert = BertModel.from_pretrained('bert-base-uncased')
        self.projection = nn.Linear(768, embed_dim)  # Project to shared latent space

    def forward(self, input_ids, attention_mask):
        features = self.bert(input_ids=input_ids, attention_mask=attention_mask).last_hidden_state[:, 0, :]
        return self.projection(features)
```

```python
class CustomDataset(Dataset):
    def __init__(self, image_paths, caption_paths, tokenizer, transform):
        self.image_paths = image_paths
        self.caption_paths = caption_paths
        self.tokenizer = tokenizer
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load and preprocess image
        image = Image.open(self.image_paths[idx]).convert("RGB")
        image = self.transform(image)

        # Load and tokenize caption
        with open(self.caption_paths[idx], 'r') as file:
            caption = file.read().strip()
        encoded = self.tokenizer(caption, padding="max_length", truncation=True, max_length=30, return_tensors="pt")

        return image, encoded['input_ids'].squeeze(0), encoded['attention_mask'].squeeze(0)
```

```python
class CLIPLoss(nn.Module):
    def __init__(self, temperature=0.1):
        super(CLIPLoss, self).__init__()
        self.temperature = temperature

    def forward(self, image_embeddings, text_embeddings):
        image_embeddings = nn.functional.normalize(image_embeddings, dim=1)
        text_embeddings = nn.functional.normalize(text_embeddings, dim=1)

        logits = torch.matmul(image_embeddings, text_embeddings.T) / self.temperature
        labels = torch.arange(len(image_embeddings)).to(image_embeddings.device)

        loss = nn.CrossEntropyLoss()
        return loss(logits, labels) + loss(logits.T, labels)
```

```python
# Define tokenizer and transforms
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Create dataset and dataloader
dataset = CustomDataset(image_paths, caption_paths, tokenizer, transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
# Initialize models, optimizer, and loss
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
image_encoder = ImageEncoder(embed_dim=256).to(device)
text_encoder = TextEncoder(embed_dim=256).to(device)
criterion = CLIPLoss().to(device)
optimizer = optim.Adam(list(image_encoder.parameters()) + list(text_encoder.parameters()), lr=1e-4)

# Training loop
for epoch in range(85):
    image_encoder.train()
    text_encoder.train()

    total_loss = 0
    for images, input_ids, attention_mask in dataloader:
        images = images.to(device)
        input_ids = input_ids.to(device)
        attention_mask = attention_mask.to(device)

        image_embeddings = image_encoder(images)
        text_embeddings = text_encoder(input_ids, attention_mask)
        loss = criterion(image_embeddings, text_embeddings)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch [{epoch+1}/100], Loss: {total_loss/len(dataloader):.4f}")
```
